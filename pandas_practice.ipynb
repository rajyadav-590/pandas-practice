```python
import pandas as pd
df = pd.read_csv("data/data.csv")
print(df.to_string())
```

```python
print(df.info())
```

```python
# empty cell handling
# 1.deleting whole row which contain null or nan value
df.dropna(inplace=True)
print(df.to_string())
```

```python
# 2.replacing specifiq column value
df.fillna({"Calories":132},inplace=True)
print(df.to_string())

# 3.for one value in place of all na use below commented statement
#df.fillna(132,inplace=True)
#print(df.to_string())
```

```python
# 4. filligng using mean ,mode,median
x=df["Calories"].mean()
#x=df["Calories"].median()
#x=df["Calories"].mode()[0]
df.fillna({"Calories":x},inplace=True)
```

```python
print(df.to_string())
```

```python
# changing format of date
df.columns
```

```python
df["Date"]=pd.to_datetime(df["Date"],format="mixed")
```

```python
df.dropna(inplace=True)
```

```python
print(df.to_string())
```

```python
# removing duplicates
# first finding duplicates are there or not
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.to_string())
```

```python
# replacing value using condition
for i in df.index:
  if df.loc[i,"Duration"]>120:
    df.loc[i,"Duration"]=120

```
